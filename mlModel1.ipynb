# ML Model 1
CNN Residual (Artifact Refiner)

1. Input parameters
- Input: `residual(t) = raw_ecg(t) - dsp_filtered_ecg(t)`
Residual contains motion artefacts, muscle bursts, leftovers from baseline.
- Input shape: `[r1, r2, r3, ..., r720]`, considering a sampling rate of 360Hz, with a window of 2s.

2. Preprocessing:
- Mean removal: `r = r - mean(r)`
- Normalization: `r = r / max(|r|)`

3. Target:
- Input to residual, target to true artefact
- Residual -> Conv1D -> ReLU -> Conv1D -> Residual skip connection(residual + CNN output) -> Refined artefact output
